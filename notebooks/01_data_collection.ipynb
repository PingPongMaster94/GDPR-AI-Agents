```python
import requests
from bs4 import BeautifulSoup
from pathlib import Path
import time


urls = [
    # --- Technology / Platforms ---
    "https://wordpress.org/about/privacy/",
    "https://openai.com/privacy", 
    "https://anthropic.com/privacy",
    "https://huggingface.co/privacy",
    "https://notion.so/privacy",
    "https://canva.com/policies/privacy-policy",
    "https://monday.com/privacy-policy",
    "https://dropcontact.com/privacy-policy",
    "https://zapier.com/privacy/",
    "https://www.cloudflare.com/en-gb/privacypolicy/",
    "https://www.spotify.com/uk/legal/privacy-policy/",
    "https://www.google.com/intl/en/policies/privacy/",
    "https://about.facebook.com/privacy/",
    "https://twitter.com/en/privacy",
    "https://www.linkedin.com/legal/privacy-policy",
    "https://help.instagram.com/519522125107875",
    "https://www.microsoft.com/en-us/privacy",
    "https://www.apple.com/legal/privacy/en-ww/",
    "https://slack.com/trust/privacy/privacy-policy",
    "https://zoom.us/privacy",
    "https://www.dropbox.com/privacy",
    "https://www.adobe.com/privacy/policy.html",
    "https://www.salesforce.com/company/privacy/",
    "https://www.atlassian.com/legal/privacy-policy",
    "https://www.trello.com/privacy",
    "https://www.airtable.com/privacy",
    "https://www.notion.so/privacy",
    "https://asana.com/terms#privacy-policy",
    "https://www.figma.com/privacy/",
    "https://www.github.com/privacy",
    "https://gitlab.com/-/legal/privacy-policy",
    "https://www.bitbucket.org/product/privacy-policy",

    # --- Cloud / IT Infrastructure ---
    "https://aws.amazon.com/privacy/",
    "https://azure.microsoft.com/en-us/support/legal/",
    "https://cloud.google.com/privacy",
    "https://www.digitalocean.com/legal/privacy-policy",
    "https://www.linode.com/legal-privacy/",
    "https://www.heroku.com/policy/privacy",
    "https://vercel.com/legal/privacy-policy",
    "https://www.netlify.com/privacy/",
    "https://www.fastly.com/privacy",
    "https://www.cloudflare.com/en-gb/privacypolicy/",
    "https://www.ibm.com/privacy/us/en/",
    "https://www.oracle.com/legal/privacy/",
    "https://www.sap.com/about/legal/privacy.html",

    # --- Finance / Fintech / Banking ---
    "https://www.paypal.com/us/legalhub/privacy-full",
    "https://stripe.com/privacy",
    "https://revolut.com/legal/privacy-policy",
    "https://www.visa.co.uk/legal/privacy-policy.html",
    "https://www.mastercard.us/en-us/vision/who-we-are/privacy.html",
    "https://wise.com/legal/privacy-policy",
    "https://www.coinbase.com/legal/privacy",
    "https://www.binance.com/en/privacy",
    "https://www.monzo.com/legal/privacy-notice/",
    "https://www.hsbc.com/privacy-notice",
    "https://www.allianz.com/en/privacy.html",
    "https://www.metlife.com/privacy/",
    "https://www.aig.com/privacy-policy",
    "https://www.chubb.com/privacy-policy.html",
    "https://www.intuit.com/privacy/",
    "https://www.americanexpress.com/en-us/company/legal/privacy-center/",
    "https://www.barclays.co.uk/important-information/privacy-policy/",
    "https://www.santander.co.uk/personal/support/customer-support/legal-information/privacy-statement",
    "https://www.lloydsbank.com/legal/privacy.html",
    "https://www.n26.com/en-eu/privacy-policy",
    "https://www.deutsche-bank.de/dam/deutschebank/de/resources/pdf/Privacy-Notice.pdf",

    # --- Retail / E-commerce ---
    "https://www.amazon.co.uk/gp/help/customer/display.html?nodeId=GX7NJQ4ZB8MHFRNJ",
    "https://www.ebay.com/help/policies/member-behaviour-policies/user-privacy-notice-privacy-policy?id=4260",
    "https://www.shopify.com/legal/privacy",
    "https://www.etsy.com/legal/privacy",
    "https://www.zalando.co.uk/privacy-policy",
    "https://www.hm.com/privacy",
    "https://www.nike.com/privacy-policy",
    "https://www.adidas-group.com/en/datenschutz/",
    "https://www.apple.com/legal/privacy/en-ww/",
    "https://www.tesco.com/help/privacy-and-cookies/",
    "https://www.carrefour.com/en/privacy-policy",
    "https://www.walmart.com/help/article/privacy-policy/0fdf0593a1774e84a7caa9cf2b32b82c",

    # --- Healthcare / Wellbeing ---
    "https://www.fitbit.com/global/legal/privacy-policy",
    "https://www.myfitnesspal.com/privacy-policy",
    "https://www.webmd.com/about-webmd-policies/about-privacy-policy",
    "https://www.healthline.com/privacy-policy",
    "https://www.pfizer.com/privacy",
    "https://www.novartis.com/privacy-policy",
    "https://www.roche.com/privacy_policy",
    "https://www.gsk.com/en-gb/privacy-notice/",
    "https://www.bayer.com/en/privacy-statement",
    "https://www.jnj.com/privacy-policy",
    "https://www.nhs.uk/our-policies/privacy-policy/",
    "https://www.mayoclinic.org/about-this-site/privacy-policy",
    "https://www.cdc.gov/other/privacy.html",
    "https://www.who.int/about/policies/privacy",
    "https://www.clevelandclinic.org/PrivacyPolicy",
    "https://www.heart.org/en/about-us/statements-and-policies/privacy-policy",

    # --- Education / Research / Learning ---
    "https://www.coursera.org/about/privacy",
    "https://www.edx.org/edx-privacy-policy",
    "https://www.khanacademy.org/about/privacy-policy",
    "https://www.harvard.edu/privacy-statement/",
    "https://www.ox.ac.uk/privacy-policy",
    "https://www.cam.ac.uk/about-this-site/privacy-policy",
    "https://www.mit.edu/privacy/",
    "https://www.ucl.ac.uk/legal-services/privacy",
    "https://www.ed.ac.uk/records-management/what-we-do/privacy-notice",
    "https://www.tcd.ie/privacy/",
    "https://www.ethz.ch/en/privacy-policy.html",
    "https://www.stanford.edu/site/privacy/",
    "https://www.yale.edu/privacy-policy",
    "https://www.lse.ac.uk/privacy",

    # --- Media / Communication ---
    "https://www.bbc.co.uk/privacy/",
    "https://edition.cnn.com/privacy",
    "https://www.nytimes.com/privacy/privacy-policy",
    "https://www.theguardian.com/info/privacy",
    "https://www.reuters.com/privacy-policy/",
    "https://www.aljazeera.com/privacy-policy/",
    "https://www.euronews.com/privacy-policy",
    "https://www.netflix.com/privacy",
    "https://www.hulu.com/privacy",
    "https://www.disneyplus.com/legal/privacy-policy",
    "https://www.spotify.com/uk/legal/privacy-policy/",
    "https://www.tiktok.com/legal/privacy-policy",
    "https://www.dw.com/en/privacy-policy/a-15778876",
    "https://www.politico.eu/privacy-policy/",
    "https://www.axios.com/privacy-policy",

    # --- Travel / Services ---
    "https://www.booking.com/content/privacy.en-gb.html",
    "https://www.airbnb.com/help/article/2855/privacy-policy",
    "https://www.uber.com/legal/en/document/?name=privacy-notice&country=portugal&lang=en",
    "https://www.lyft.com/privacy",
    "https://www.marriott.com/about/privacy.mi",
    "https://www.hilton.com/en/p/policies/privacy-policy/",
    "https://www.accor.com/security-cookies-policy/index.en.shtml",
    "https://www.lufthansa.com/pt/en/privacy-statement",
    "https://www.emirates.com/english/privacy-policy/",
    "https://www.skyscanner.net/legal/privacy-policy",
    "https://www.expedia.com/lp/lg-privacypolicy",
    "https://www.tripadvisor.com/pages/privacy.html",
    "https://www.kayak.com/privacy",
    "https://www.easyjet.com/en/policy/privacy",
    "https://www.ryanair.com/ie/en/corporate/privacy-policy",

    # --- Government / Public Sector ---
    "https://www.gov.uk/help/privacy-notice",
    "https://ec.europa.eu/info/privacy-policy_en",
    "https://www.nasa.gov/privacy/",
    "https://www.whitehouse.gov/privacy/",
    "https://www.canada.ca/en/transparency/privacy.html",
    "https://www.australia.gov.au/privacy",
    "https://www.un.org/en/about-us/privacy-notice",
    "https://www.oecd.org/privacy/",
    "https://www.imf.org/en/About/terms/privacy-policy",
    "https://www.wto.org/english/privacy_e.htm",



    "https://www.coca-cola.com/privacy-policy",
    "https://www.pepsico.com/legal/privacy",
    "https://www.unilever.com/privacy-notice/",
    "https://www.loreal.com/en/privacy-policy/",
    "https://www.p-and-g.com/privacy",
    "https://www.sony.com/en/privacy/"


     "https://legal.hubspot.com/privacy-policy",
    "https://discord.com/privacy",
    "https://www.reddit.com/policies/privacy-policy",
    "https://www.mozilla.org/en-US/privacy/websites/",
    "https://automattic.com/privacy/",
    "https://miro.com/legal/privacy-policy/",
    "https://www.intercom.com/legal/privacy",
    "https://calendly.com/legal/privacy-notice",
    "https://proton.me/legal/privacy",
    "https://1password.com/legal/privacy",
    "https://www.docusign.com/privacy",
    "https://www.deepl.com/en/privacy",
    "https://www.grammarly.com/privacy-policy",
    "https://www.zendesk.com/company/agreements-and-terms/privacy-notice/",
    "https://www.adyen.com/privacy-policy",
    "https://plaid.com/legal/#privacy-policy",
    "https://www.sofi.com/privacy-policies/",
    "https://www.checkout.com/legal/privacy-policy",
    "https://www.brex.com/legal/privacy",
    "https://www.afterpay.com/en-US/privacy-policy",
    "https://www.payoneer.com/legal/privacy-policy/",
    "https://www.target.com/c/target-privacy-policy/-/N-4sr7p",
    "https://www.bestbuy.com/site/help-topics/privacy-policy/pcmcat204400050062.c?id=pcmcat204400050062",
    "https://www.asos.com/privacy-policy/",
    "https://www.chewy.com/app/content/privacy",
    "https://www.costco.com/privacy-policy.html",
    "https://privacy-hub.sainsburys.co.uk/privacy-policy/",
    "https://www.udemy.com/terms/privacy/",
    "https://www.futurelearn.com/info/terms/privacy-policy",
    "https://www.duolingo.com/privacy",
    "https://www.mheducation.com/about-us/trust-center/privacy",
    "https://www.instructure.com/policies/product-privacy-policy",
    "https://qualifications.pearson.com/en/privacy-policy.html",
    "https://loginstatic.pearson.com/html/PearsonPrivacyPolicy.html",
    "https://www.nbcuniversalprivacy.com/privacy",
    "https://www.washingtonpost.com/privacy-policy/",
    "https://www.theatlantic.com/privacy-policy/",
    "https://www.economistgroup.com/privacy-policy",
    "https://www.zocdoc.com/about/privacypolicy/",
    "https://ada.com/privacy-policy/",
    "https://www.nuffieldhealth.com/pages/privacy",
    "https://www.unitedhealthgroup.com/privacy.html",
    "https://www.23andme.com/en-int/legal/privacy/",
    "https://www.headspace.com/privacy-policy",
    "https://ico.org.uk/global/privacy-notice/",
    "https://www.nih.gov/web-policies-notices",
    "https://www.uspto.gov/privacy-policy",
    "https://www.parliament.uk/site-information/data-protection/",
    "https://www.federalreserve.gov/privacy.htm",
    "https://www.nist.gov/admr/oism/site-privacy"




]


SAVE_DIR = Path("/Users/davidj.silva/Desktop/Data Science/Thesis/2. Project/data/raw_policies")
SAVE_DIR.mkdir(parents=True, exist_ok=True)

headers = {"User-Agent": "Mozilla/5.0"}
session = requests.Session()

for url in urls:
    domain = url.split("/")[2].replace(".", "_")
    file_path = SAVE_DIR / f"{domain}_privacy.txt"

    for attempt in range(3):          # retry a few times
        try:
            r = session.get(url, headers=headers, timeout=20, stream=True)
            r.raise_for_status()
            r.encoding = r.apparent_encoding
            text = r.text
            soup = BeautifulSoup(text, "html.parser")
            body_text = soup.get_text(separator="\n", strip=True)
            file_path.write_text(body_text, encoding="utf-8")
            print(f"✅ Saved: {file_path.name}")
            break                      # success → exit retry loop

        except (requests.exceptions.ChunkedEncodingError,
                requests.exceptions.ConnectionError,
                requests.exceptions.ReadTimeout) as e:
            print(f"⚠️ Attempt {attempt+1}/3 failed for {domain}: {e}")
            time.sleep(3)              # wait before retry
        except Exception as e:
            print(f"❌ Skipped {domain}: {e}")
            break
```
